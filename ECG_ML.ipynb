# Train Test Split

```python
X=df.drop('classification',axis=1)
y=df["classification"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
```

```python
Scaled_X_train=scaler.transform(X_train)
Scaled_X_test=scaler.transform(X_test)
```

## Fitting Model Using SVM

```python
from sklearn.svm import SVC
model=SVC()
model.fit(Scaled_X_train,y_train)
```

```python
pred=model.predict(Scaled_X_test)
```

```python
pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix,accuracy_score
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
plot_confusion_matrix(model,Scaled_X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.001,0.01,0.1,0.5,1],
             'kernel':['linear','rbf','poly'],
              'gamma':['scale','auto']
               }
```

```python
grid_model=GridSearchCV(model,param_grid)
```

```python
grid_model.fit(Scaled_X_train,y_train)
```

```python
grid_model.best_params_
```

```python
grid_pred=grid_model.predict(Scaled_X_test)
```

```python
accuracy_score(y_test,grid_pred)
```

```python
plot_confusion_matrix(grid_model,Scaled_X_test,y_test)
```

## Fitting Model Using RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
rfc.fit(Scaled_X_train,y_train)
```

```python
rfc_pred=rfc.predict(Scaled_X_test)
```

```python
confusion_matrix(y_test,rfc_pred)
```

```python
accuracy_score(y_test,rfc_pred)
```

```python
plot_confusion_matrix(rfc,Scaled_X_test,y_test)
```

```python
n_estimators=[64,100,128,200]
max_features=[2,3,4]
```

```python
param_grid={'n_estimators':n_estimators,'max_features':max_features}
```

```python
grid_rfc=GridSearchCV(rfc,param_grid)
```

```python
grid_rfc.best_params_
```

```python
grid_rfc_pred=grid_rfc.predict(Scaled_X_test)
```

```python
accuracy_score(y_test,grid_rfc_pred)
```

```python
plot_confusion_matrix(grid_rfc,Scaled_X_test,y_test)
```
